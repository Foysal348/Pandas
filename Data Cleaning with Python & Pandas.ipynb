```python
import pandas as pd 
import numpy as np 
df=pd.read_csv("Pandas_students_data.csv")
df.head(5)
```

```python
df.info()
```

```python
#Disply Duplicate
df.duplicated()
```

```python
#Delete Duplicate value
df=df.drop_duplicates()
df.head(5)
```

```python
df.isnull().sum()
```

```python
#name Handling
df["name"]=df["name"].str.lower().str.title()
df.head()
```

```python
#handling age
#df["age"].unique()
mean_age=df["age"].mean().round(0)
df["age"]=df["age"].fillna(mean_age)
df.head(5)
```

```python
#Gender Handling 
#df["gender"].unique()
df["gender"]=df['gender'].str.title()
df.head(5)
```

```python
#Grade Handling 
#df["grade"].unique()
df["grade"]=df["grade"].replace({"10th":10,"11th":11})
df.head(5)
```

```python
#math score Handling 
#df["math_score"].unique()
mean_math=df["math_score"].mean()
df["math_score"]=df["math_score"].fillna(mean_math)
df.head()
```

```python
#english score Handling 
#df["english_score"].unique()
mean_english=df["math_score"].mean()
#pd.to_numeric(df["english_score"],errors="coerce")# alpha to numeric(nan)
df["english_score"]=df["english_score"].fillna(mean_english)
df["english_score"]=df["english_score"].replace({"missing":mean_english})
```

```python
#remarks Handling 
df["remarks"].unique()
#df.loc[:,"remarks"]==df["remarks"].str.lower().str.title()
dic={
    'excellent':"Excellent", 'GOOD':"Good", 'needs improvement':"Need Improvement", 'average':"Average", 'Good':"Good",
       'poor':"Poor", 'Average':"Average", 'good student':"Good Student"
}
df["remarks"]=df["remarks"].replace(dic)
df.head()
```
