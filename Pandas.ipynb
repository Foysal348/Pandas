```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob"],
    "Math": [90, 80],
    "Science": [85, 88]
})

print(df)
```

```python
pd.melt(df)
```

```python
df_melted = pd.melt(df, id_vars=["Name"], var_name="Subject", value_name="Score")

#print(df_melted)
df_melted
```

**Pivot**

```python
#reshapes data from long format to wide format, which is the inverse of melt().
df_melted.pivot(index="Name",columns="Subject")
```

```python
df_melted.pivot_table(index="Subject", values="Score", aggfunc="mean")
#df_melted.pivot_table(index="Subject", values="Score", aggfunc="sum")
```
